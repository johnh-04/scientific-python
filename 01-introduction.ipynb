## 2. Arithmetic Operators

```python
distance = 1000
time = 3

# Arithmetic operations (+, -, *, /, //, %, **)
print("Arithmetic operations:")
print("Addition (+):", distance + time)
print("Subtraction (-):", distance - time)
print("Multiplication (*):", distance * time)
print("Standard division (/):", distance / time)
print("Integer division (//):", distance // time)
print("Modulo or Remainder (%):", distance % time)
print("Power (**):", 5 ** 2)

# Formatting with round() function
print("\nFormatting with round() function:")
print("Rounding to 2 decimal places:", round(distance / time, 2))
```

## 3. Conditional Structure and Loops (with Enumerate)

```python
# If-Elif-Else
print("If-Elif-Else:")
num = 10
if num > 0 and num < 20:
    print("Positive and less than 20")
elif num < 0 or num > 20:
    print("Negative or greater than 20")
else:
    print("Zero")

# If with not
print("\nIf with not:")
if not num == 0:
    print("Not zero")

# While loop
print("\nWhile loop:")
x = 0
while x < 3:
    print(f"While index: {x}")
    x += 1

# While True with break
print("\nWhile True with break:")
counter = 0
while True:
    print(f"Counter: {counter}")
    counter += 1
    if counter >= 3:
        break

# For loop
print("\nFor loop:")
for i in range(3):
    print(f"For index: {i}")

# Enumerate function
print("\nEnumerate function:")
protocols = ["I2C", "SPI", "UART"]
for index, protocol in enumerate(protocols):
    print(f"Protocol n° {index}: {protocol}")
```

## 5. Collective Data Structures

### List

List: ordered, mutable, allows duplicates, heterogeneous

```python
# List
mixed_list = [1, "two", 3.0, True]

# List elements access
print("List elements access:")
print("List:", mixed_list)
print("First element:", mixed_list[0])
print("Fourth element:", mixed_list[3])
print("Last element:", mixed_list[-1])  

# List iteration
print("\nList iteration:")
for item in mixed_list:
    print(item)

# List slicing
print("\nList slicing:")
print("Sliced list (1:3):", mixed_list[1:3])
print("Sliced list (:2):", mixed_list[:2])
print("Sliced list (2:):", mixed_list[2:])

# List slicing with step
print("\nList slicing with step:")
print("Every second element:", mixed_list[::2])
print("Reversed list:", mixed_list[::-1])

# List concatenation
print("\nList concatenation:")
print("Concatenated list:", mixed_list + [5, "six"])

# List properties and methods
print("\nList properties and methods:")
print("Length of the list:", len(mixed_list))

# Membership test
print("\nMembership test:")
print("Membership test ('two' in mixed_list):", "two" in mixed_list)
print("Membership test ('three' not in mixed_list):", "three" not in mixed_list)

# Membership test with numbers
print("\nMembership test with numbers:")
nums = [1, 2, 3, 4, 5]
if 10 not in nums:
    print("10 is not in the list")

# Modifying lists
print("\nModifying lists:")

mixed_list[1] = "changed"
print("Modified list:", mixed_list)

mixed_list.append("new")
print("Appended element:", mixed_list)

mixed_list.insert(2, "inserted")
print("Inserted element:", mixed_list)

mixed_list.extend([7, "eight"])
print("Extended list:", mixed_list)

mixed_list.remove("changed")
print("Removed element:", mixed_list)

last_element = mixed_list.pop()
print("Popped element:", last_element)
print("List after popping:", mixed_list)

del mixed_list[0]
print("List after deleting first element:", mixed_list)
mixed_list.clear()
print("Cleared list:", mixed_list)

# Sorting lists
print("\nSorting lists:")
nums = [5, 2, 9, 1]
nums.sort()
print("Sorted list:", nums)
nums.sort(reverse=True)
print("Sorted in reverse:", nums)

# List comprehensions
print("\nList comprehensions:")

squares = [x**2 for x in range(5)]
print("List comprehension (Squares):", squares)

sats_database = ["NOAA 15", "NOAA 18", "NOAA 20", "NOAA 21"]
active_sats = [sat for sat in sats_database if "20" in sat or "21" in sat]
print("List comprehension (Filtered Satellites):", active_sats)
```

### Set

Set: unordered, mutable, no duplicates, heterogeneous

```python
# Set
num_set = {1, 2, 2, 3, 4, 5}

# Set elements access
print("Set elements access:")
print("Set:", num_set)

# Set iteration
print("\nSet iteration:")
for item in num_set:
    print(item)

# Set properties and methods
print("\nSet properties and methods:")
print("Length of set:", len(num_set))

# Modifying sets
print("\nModifying sets:")

num_set.add(6)
print("Set after adding 6:", num_set)

num_set.remove(3)
print("Set after removing 3:", num_set)

num_set.discard(10)
print("Set after discarding 10:", num_set)

A = {1, 2, 3}
B = {3, 4, 5}

# Membership test
print("\nMembership test:")
print("3 in A:", 3 in A)
print("10 not in A:", 10 not in A)

# Set operations
print("\nSet operations:")

print("A | B:", A | B)
print("A union B:", A.union(B))

print("A & B:", A & B)
print("A intersection B:", A.intersection(B))

print("A - B:", A - B)
print("A difference B:", A.difference(B))

print("A ^ B:", A ^ B)
print("A symmetric difference B:", A.symmetric_difference(B))
```

### Dictionary

Dictionary: unordered, mutable, unique keys, heterogeneous values

```python
# Dictionary
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Alternative way to create a dictionary
person = dict(name="Alice", age=30, city="New York")

# Dictionary elements access
print("Dictionary elements access:")
print("Dictionary:", person)
print("Name:", person["name"])
print("Age:", person["age"])
print("City:", person["city"])

# Dictionary access with get() method
print("\nDictionary access with get() method:")
print("Age (using get):", person.get("age"))
print("Country (using get):", person.get("country", "Unknown"))

# Dictionary iteration
print("\nDictionary iteration:")

print("Iterating through keys:")
for key in person:
    print("Key:", key)

print("Iterating through values:")
for value in person.values():
    print("Value:", value)

print("Iterating through key-value pairs:")
for key, value in person.items():
    print(f"{key}: {value}")

# Methods and properties
print("\nMethods and properties:")
print("Number of key-value pairs:", len(person))

# Membership test
print("\nMembership test:")
print("name" in person)
print("city" not in person)

# Dictionary views
print("\nDictionary views:")
print("Keys:", person.keys())
print("Values:", person.values())
print("Items:", person.items())

# Modifying dictionaries
print("\nModifying dictionaries:")

person["age"] = 31
print("Updated Dictionary:", person)

person["country"] = "USA"
print("Added Key-Value Pair:", person)

person.pop("age")
print("After Popping 'age':", person)

person.popitem()
print("After Popping Last Item:", person)

del person["city"]
print("After Deleting 'city':", person)

person.clear()
print("After Clearing Dictionary:", person)

# Nested dictionaries
print("\nNested dictionaries:")
students = {
    "Alice": {"age": 30, "grade": "A"},
    "Bob": {"age": 25, "grade": "B"}
}

print("Nested Dictionary:", students)
print("Alice's grade:", students["Alice"]["grade"])
print("Bob's age:", students["Bob"]["age"])
```

## 6. Functions and Recursion

```python
# Function
def sum(a, b):
    return a + b

print("Function:")
print(sum(3, 5))

# Default parameter values
def greet(name="World"):
    return f"Hello, {name}!"

print("\nDefault parameter values:")
print(greet())
print(greet("Alice"))

# Global vs Local variables
x = 10          # Global variable
def modify_global():
    global x    # Declare x as global to modify it
    y = 5       # Local variable
    x += y      # Modify global variable x

print("\nGlobal vs Local variables:")
modify_global()
print("Modified global variable x:", x)

# Nested functions
def outer_function():
    def inner_function():
        return "Hello from the inner function!"
    return inner_function()

print("\nNested functions:")
print(outer_function())

# Recursion (factorial)
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
print("\nRecursion (factorial):")
print("Factorial of 5:", factorial(5))

# Recursion (Fibonacci)
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
    
print("\nRecursion (Fibonacci):")
print("Fibonacci of 10:", fibonacci(10))
```
