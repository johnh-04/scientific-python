````python
import pandas as pd
````

````python
# Series
series = pd.Series([1, 2, 3, 4, 5])
print("Series:")
print(series)

# DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [30, 25, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data)
print("\nDataFrame:")
print(df)
````

## 2. Creating a DataFrame from a CSV file

````python
df_csv = pd.read_csv("./assets/data.csv")

# Exploring DataFrame (first 5 rows by default)
print("DataFrame from CSV:")
print(df_csv.head())
display(df_csv.head())

# Read CSV with a different separator
df = pd.read_csv("./assets/data.csv", sep=",")
print("\nDataFrame with custom separator:")
print(df.head())

# Read CSV without header
df = pd.read_csv("./assets/data.csv", header=None)
print("\nDataFrame without header:")
print(df.head())

# Read CSV with custom column names
df = pd.read_csv("./assets/data.csv", names=["Column1", "Column2", "Column3"])
print("\nDataFrame with custom column names:")
print(df.head())

# Read CSV and set a column as index
df = pd.read_csv("./assets/data.csv", index_col="Name")
print("\nDataFrame with index column:")
print(df.head())

# Read only specific columns from CSV
df = pd.read_csv("./assets/data.csv", usecols=["Name", "Age", "City"])
print("\nDataFrame with specific columns:")
print(df.head())

# Skip the first row of the CSV file
df = pd.read_csv("./assets/data.csv", skiprows=1)
print("\nDataFrame with skipped row:")
print(df.head())

# Read only the first 3 rows of the CSV file
df = pd.read_csv("./assets/data.csv", nrows=3)
print("\nDataFrame with first 3 rows:")
print(df.head())

# Treat specific values as NaN
df = pd.read_csv("./assets/data.csv", na_values=["NA", "N/A"])
print("\nDataFrame with NaN values:")
print(df.head())

# Parse specific columns as dates
df = pd.read_csv("./assets/data.csv", parse_dates=["Date"])
print("\nDataFrame with parsed dates:")
print(df.head())

# Specify data types for columns
df = pd.read_csv("./assets/data.csv", dtype={"Name": str, "Age": int, "City": str})
print("\nDataFrame with specified data types:")
print(df.head())

# Specify encoding when reading CSV file
df = pd.read_csv("./assets/data.csv", encoding="utf-8")
print("\nDataFrame with specified encoding:")
print(df.head())
````

## 3. Exploring the DataFrame

````python
# DataFrame Information
print("Info:", df.info())

# Statistical Summary
print("\nDescribe:", df.describe())

# DataFrame Shape
print("\nShape:", df.shape)
````

## 4. Accessing and Filtering Data

````python
# Accessing data (Series)
print("Age column:")
print(df["Age"])
print("\nName and City columns:")
print(df[["Name", "City"]])

# Accessing data (DataFrame)
print("\nFirst row:")
print(df.iloc[0])
print("\nFirst two rows:")
print(df.iloc[0:2])

# Accessing data by label
print("\nName in first row:")
print(df.loc[0, "Name"])

# Sorting data (DataFrame)
print("\nSorted by Age:")
print(df.sort_values(by="Age"))

# Filtering data (DataFrame)
print("\nAge > 30:")
print(df[df["Age"] > 30])
print("\nCity == New York:")
print(df[df["City"] == "New York"])
````

## 5. Modifying and Saving Data

````python
# Modifying data (DataFrame)
df["Age"] = df["Age"] + 1
print("Incremented Age:")
print(df)

df.drop(columns=["City"], inplace=True)
print("\nDropped City column:")
print(df)

df.loc[df["Name"] == "Alice", "Age"] = 32
print("\nUpdated Alice's Age:")
print(df)

df.drop(index=0, inplace=True)
print("\nDropped first row:")
print(df)

# Saving data
df.to_csv("./assets/updated_data.csv", index=False)
print("\nDataFrame successfully saved to 'updated_data.csv'")
````
