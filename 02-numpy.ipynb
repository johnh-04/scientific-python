# NumPy Guide
Numpy is a powerful library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

````python
import numpy as np
````

````python
# Creating a numpy array
list_vals = [1, 2, 3, 4, 5]
arr = np.array(list_vals)
print("Creating a numpy array:")
print("Array:", arr)
print("Type:", type(arr))

# Creating a numpy array with a specific data type
arr_float = np.array(list_vals, dtype=float)
print("\nCreating a numpy array with a specific data type:")
print("Float Array:", arr_float)
print("Dtype:", arr_float.dtype)

# Creating a numpy matrix
list_2d = [[1, 2, 3], [4, 5, 6]]
matrix = np.array(list_2d)
print("\nCreating a numpy matrix:")
print("Matrix:\n", matrix)  
print("Type:", type(matrix)) 
````

## 2. Special Arrays and Ranges

````python
# Zeros Matrix
zeros_array = np.zeros((3, 3))
print("Zeros Matrix:")
print("Zeros:\n", zeros_array)

# Ones Matrix
ones_array = np.ones((2, 4))
print("\nOnes Matrix:")
print("Ones:\n", ones_array)

# Identity Matrix
eye_array = np.eye(4)
print("\nIdentity Matrix:")
print("Identity:\n", eye_array)

# Full Matrix
full_array = np.full((2, 3), 7)
print("\nFull Matrix:")
print("Full:\n", full_array)

# Ranges
range_array = np.arange(0, 10, 2)
print("\nRanges:")
print("Arange:", range_array)
linspace_array = np.linspace(0, 1, 5)
print("\nLinspace:")
print("Linspace:", linspace_array)
````

## 3. Array Properties & Random Arrays

````python
# Array properties
print("Array properties:")
print("Shape:", matrix.shape)
print("Dimensions:", matrix.ndim)
print("Size:", matrix.size)
print("Dtype:", matrix.dtype)

# Random arrays
print("\nRandom arrays:")

np.random.seed(42) # Set seed for reproducibility 
random_array = np.random.rand(3, 3)
print("Random (0 to 1):\n", random_array)  

random_int_array = np.random.randint(0, 10, (4, 4))
print("Random Integers:\n", random_int_array)
````

## 4. Mathematical, Matrix Operations & Linear Algebra

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise operations
print("Element-wise operations:")
print("Add:", a + b)
print("Sub:", a - b)
print("Mult:", a * b)
print("Div:", a / b)
print("Power:", a ** 2)

# Matrix multiplication/Dot product
print("\nMatrix multiplication/Dot product:")
print("Dot method:", a.dot(b))
print("Np dot:", np.dot(a, b))
print("Operator @:", a @ b)

# Matrix transpose
print("\nMatrix transpose:")
print("Transpose:", a.T)

# Eigenvalues 
G = np.array([[1, 7, 2],
              [5, 4, 3],
              [1, 3, 5]])
eigenvalues = np.linalg.eigvals(G)
print("\nEigenvalues:", eigenvalues)
````

## 5. Indexing, Slicing and Boolean Masking

````python
arr = np.array([10, 20, 30, 40, 50, 60, 70])

# Indexing
print("Indexing:")
print("First:", arr[0])
print("Last:", arr[-1])

# Slicing
print("\nSlicing:")
print("Slice 1 to 3:", arr[1:4])
print("Slice to 2:", arr[:2])
print("Reverse:", arr[::-1])

# Boolean Masking (Filtering with conditions)
print("\nBoolean Masking:")

mask_and = (arr > 20) & (arr < 60)
print("Mask (20 < x < 60):", arr[mask_and])

mask_or = (arr == 10) | (arr > 50)
print("Mask (x == 10 OR x > 50):", arr[mask_or])
````

## 6. Statistical Functions

````python
data = np.array([1, 2, 3, 4, 5])

# Basic statistics
print("Basic statistics:")
print("Sum:", np.sum(data))  
print("Mean:", np.mean(data))  
print("Median:", np.median(data))  

# Standard Deviation and Variance (with degrees of freedom)
print("\nStandard Deviation and Variance (with degrees of freedom):")
print("Std Dev:", np.std(data))  
print("Std Dev (ddof=1):", np.std(data, ddof=1))  
print("Variance:", np.var(data))  
print("Variance (ddof=1):", np.var(data, ddof=1))  

# Min, Max, Percentiles, Correlation, Covariance
print("\nMin, Max, Percentiles, Correlation, Covariance:")
print("Min:", np.min(data))  
print("Max:", np.max(data))  
print("25th Percentile:", np.percentile(data, 25))  
print("75th Percentile:", np.percentile(data, 75))  
print("Correlation Coefficient:\n", np.corrcoef(data, data))  
print("Covariance:\n", np.cov(data, data)) 

# Sorting and Weighted Average
print("\nSorting and Weighted Average:")

unordered_data = np.array([543, 243, 743, 283])
print("Sorted:", np.sort(unordered_data))

weights = np.array([0.1, 0.4, 0.3, 0.2])
print("Weighted Average:", np.average(unordered_data, weights=weights))
````

## 7. Universal Functions (ufuncs)

````python
arr = np.array([0, np.pi / 2, np.pi])

# Mathematical functions
print("Mathematical functions:")
print("Sin:", np.sin(arr))  
print("Cos:", np.cos(arr))  
print("Exp:", np.exp(arr))  
print("Log (base e):", np.log(arr + 1)) 
print("Log10 (base 10):", np.log10(arr + 1)) 
print("Sqrt:", np.sqrt(arr)) 
print("Abs:", np.abs(arr - np.pi / 2))  
print("Round:", np.round(arr, decimals=2))  
print("Floor:", np.floor(arr)) 
print("Ceil:", np.ceil(arr))  
print("Sign:", np.sign(arr - np.pi / 2)) 
print("Where:", np.where(arr > np.pi / 2, "greater", "not greater"))  
print("Unique:", np.unique(arr))
````

## 8. Write and Read from Files

````python
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Basic TXT Saving/Loading
np.savetxt("./assets/gen_array.txt", arr, delimiter=",")
loaded_arr = np.loadtxt("./assets/gen_array.txt", delimiter=",")
print("Basic TXT Saving/Loading:")
print("Loaded TXT:\n", loaded_arr)  

# NPY Saving/Loading (Binary, much faster)
np.save("./assets/gen_array.npy", arr)
loaded_arr_npy = np.load("./assets/gen_array.npy")
print("\nNPY Saving/Loading (Binary, much faster):")
print("Loaded NPY:\n", loaded_arr_npy)  

# Advanced TXT Loading (np.genfromtxt)
# Allows skipping specific lines (e.g., text headers/footers) and selecting specific columns
data_from_txt = np.genfromtxt('./assets/read_array.txt', skip_header=1, skip_footer=0, usecols=[0, 2], delimiter=",")
print("\nAdvanced TXT Loading (np.genfromtxt):")
print("Loaded with genfromtxt:\n", data_from_txt)
````
